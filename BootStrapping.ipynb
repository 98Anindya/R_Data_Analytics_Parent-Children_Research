```r
library(boot)

# Load your data
data <- read.csv("Complete_Data_Modified.csv")

# Number of rows in the new data file
num_rows_new_data <- 2000

# Set the seed for reproducibility
set.seed(123)

# Create a new data frame by sampling with replacement from the original data
new_data <- data[sample(nrow(data), num_rows_new_data, replace = TRUE), ]

# Write the new data frame to a CSV file
write.csv(new_data, "new_data.csv", row.names = FALSE)

# Display a message indicating success
cat("New data file 'new_data.csv' created with", num_rows_new_data, "rows.\n")
```

```r
library(boot)

# Load your data
data <- read.csv("Complete_Data_Modified.csv")

# Define your variables
independent_variables <- c('BFI..E.', 'BFI..A.', 'BFI..C.', 'BFI..N.', 'BFI..O.', 'EI..Self.A.', 'EI..Self.M.', 'EI..Social.A.', 'EI..RM.', 'Total')
dependent_variables <- c('Panas..', 'Panas...1', 'CBCL')

# Number of Bootstrap Samples
num_bootstrap_samples <- 1800

# Function to calculate alpha for bootstrapping
alpha_fn <- function(data, independent_variable, dependent_variable, index) {
  X <- data[[independent_variable]][index]
  Y <- data[[dependent_variable]][index]
  return((var(Y) - cov(X, Y)) / var(X) + var(Y) - 2 * cov(X, Y))
}

# Function to estimate linear regression coefficients for bootstrapping
boot_fn <- function(data, independent_variable, dependent_variable, index) {
  return(coef(lm(data[[dependent_variable]] ~ data[[independent_variable]], data = data, subset = index)))
}

# Loop through all combinations of independent and dependent variables
for (independent_variable in independent_variables) {
  for (dependent_variable in dependent_variables) {
    # Bootstrap for alpha
    boot_alpha <- boot(data, alpha_fn, independent_variable = independent_variable, dependent_variable = dependent_variable, R = num_bootstrap_samples)

    # Bootstrap for linear regression coefficients
    boot_regression <- boot(data, boot_fn, independent_variable = independent_variable, dependent_variable = dependent_variable, R = num_bootstrap_samples)

    # Display results
    cat("Results for", independent_variable, "as independent variable and", dependent_variable, "as dependent variable:\n")
    print("Alpha Estimates:")
    print(boot_alpha)

    print("Linear Regression Coefficient Estimates:")
    print(boot_regression)

    # Confidence intervals for linear regression coefficients
    ci_intercept <- boot.ci(boot_regression, index = 1, type = "perc")
    ci_slope <- boot.ci(boot_regression, index = 2, type = "perc")

    print("Confidence Intervals for Intercept:")
    print(ci_intercept)

    print("Confidence Intervals for Slope:")
    print(ci_slope)

    # Plotting
    plot(boot_regression)
  }
}
```
